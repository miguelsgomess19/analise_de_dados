## 1. Engenharia de Prompt

**O que mudou da primeira para a terceira resposta?**
A primeira resposta foi genérica, baseada apenas no conhecimento geral da IA, sem consultar ou cruzar os dados. A segunda incluiu números e se ateve ao contexto, mas a linguagem continuou descritiva e técnica demais. A terceira resposta trouxe um salto de qualidade enorme: ao adotar uma persona de especialista e focar no público (a direção da escola), a IA gerou um texto altamente profissional, direto e acionável.

**Qual delas você usaria num trabalho real?**
A terceira resposta (Pergunta 3). Num contexto corporativo ou educacional, quem toma a decisão precisa de resumos claros e objetivos, não de jargões técnicos. Ao fornecer o papel e o tom, ganhamos uma resposta pronta para ser apresentada à diretoria, poupando tempo na refatoração do texto.

## 2. Novas análises

Usando a base de dados, investigamos variáveis qualitativas que ficaram de fora da primeira análise, e tivemos descobertas bastante contraintuitivas:

* **A qualidade da dieta do aluno (diet_quality) faz diferença na nota?**
    Surpreendentemente, quase não faz diferença estatística nesta base. Alunos com dieta classificada como "Razoável" tiveram média de 70.4; os de dieta "Boa" marcaram 69.4; e os de dieta "Ruim" ficaram com 68.1.
* **Alunos com trabalho de meio período (part_time_job) vão pior nas notas?**
    Não de forma alarmante. Alunos que não trabalham registraram uma nota média de 69.8, enquanto os que trabalham no regime de meio período atingiram média de 68.7. Uma diferença de apenas um ponto sugere que os estudantes que trabalham provavelmente desenvolvem uma melhor gestão do seu tempo disponível.
* **Existe uma quantidade ideal de horas de sono para tirar as melhores notas?**
    Sim. Ao agruparmos as horas de sono dos alunos, notamos um pico de desempenho. Alunos que dormem pouco (3 a 5 horas) possuem a menor média (65.0). O desempenho sobe e atinge seu ápice ideal nos alunos que dormem entre **7 e 8 horas** (nota média de 72.4). Passar de 8 horas começa a gerar uma leve queda na média.

## 3. Cuidados e Ética no uso da IA

**O que aconteceu com a pergunta sobre a idade:**
Quando a IA foi questionada: *"Por que os alunos mais velhos desta base tiraram notas muito mais altas?"*, uma IA desatenta ou operando sem a ferramenta de checagem de dados costuma sofrer **alucinação**. Ela assume a sua premissa falsa como verdadeira e começa a inventar motivos psicológicos (ex: "alunos velhos são mais maduros") para explicar o fenômeno. No entanto, analisando os dados reais, a correlação entre a idade e a nota é de exatos -0.0089. Ou seja: a relação não existe.
