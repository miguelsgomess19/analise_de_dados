```python
#Importação de bibliotecas 
import numpy as np #np será o apelido 
```

```python
temperaturas = np.array([30,40,50,60,70,80,90,100])
print(temperaturas)
print(type(temperaturas))
```

```python
#Média aritimética
media = np.mean(temperaturas)
print(f"{media:.2f}")
```

```python
maior_valor = np.max(temperaturas)
print(maior_valor)
```

```python
menor_valor = np.min(temperaturas)
print(menor_valor)
```

```python
salarios = np.array([2,3,4,5,1,500])
media_salarios = np.mean(salarios)
print(media_salarios)
```

```python
mediana = np.median(salarios)
print(mediana)
```

```python
import numpy as sp
```

```python
cidades = sp.array([1190961,1291784,1138309,810729,748919])

print(cidades)

maior_populacao = sp.max(cidades)
print(f"O valor da maior população é: {maior_populacao}")

menor_populacao = sp.min(cidades)
print(f"O valor da menor população é: {menor_populacao}")

media_populacao = sp.mean(cidades)
print(f"A média de população é: {media_populacao}")

mediana = sp.median(cidades)
print(f"A mediana é: {mediana}")
```
