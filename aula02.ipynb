Programa que avalia as notas que o usuário do nosso aplicativo inseriu

```python
avaliacao = int(input("Digite sua avaliação: 0 - 10 estrelas"))

if avaliacao > 8: 
    print("Nota execelente!")
elif avaliacao > 6:
    print("Nota regular!")
elif avaliacao > 4:
    print("Nota ruim!")
else:
    print("Nota péssima!")
```
